```python
import pandas as pd
```

```python
vine_df = pd.read_csv("vine_table")
```

```python
vine_df.head()
```

```python
vine_total_df = vine_df[vine_df["total_votes"]>= 20]
vine_total_df.head()
```

```python
vine_ratio_df = vine_total_df[(vine_total_df["helpful_votes"]/vine_total_df["total_votes"])>.5]
vine_ratio_df
```

```python
vine_paid_df = vine_ratio_df[vine_ratio_df["vine"] == "Y"]
vine_paid_df.head()
```

```python
vine_unpaid_df = vine_ratio_df[vine_ratio_df["vine"] == "N"]
vine_unpaid_df.head()
```

```python
vine_paid_count = len(vine_paid_df)
```

```python
vine_unpaid_count = len(vine_unpaid_df)
```

```python
vine_paid_star_count = len(vine_paid_df[vine_paid_df["star_rating"]==5])
vine_paid_count
```

```python
vine_unpaid_star_count = len(vine_unpaid_df[vine_unpaid_df["star_rating"]==5])
vine_unpaid_star_count
```

```python
paid_star_percent = (vine_paid_star_count/vine_paid_count)*100
paid_star_percent
```

```python
unpaid_star_percent = (vine_unpaid_star_count/vine_unpaid_count)*100
unpaid_star_percent
```
